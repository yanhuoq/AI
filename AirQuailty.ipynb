```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('AirQualityUCI.csv')
df1=df1[~df1["CO(GT)"].isin([-200])]
df1=df1[~df1["PT08.S1(CO)"].isin([-200])]
dataset = df1.copy()
```

```python
X = dataset[["CO(GT)"]].dropna()
Y = dataset[["PT08.S1(CO)"]].dropna()
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,random_state = 0)
```

```python
def plot_fitting_line(linear_reg_model, X, y, feat):
    w = linear_reg_model.coef_
    b = linear_reg_model.intercept_
    plt.figure()   
    plt.scatter(X, y, alpha=0.5)
    plt.plot(X, w * X + b, c='red')
    plt.title(feat)  
    plt.show()
```

```python
linear_reg_model = LinearRegression()
linear_reg_model.fit(X_train, Y_train)

print("训练集准确率：{:.4f}".format(linear_reg_model.score(X_train, Y_train)))
print("测试集准确率：{:.4f}".format(linear_reg_model.score(X_test, Y_test)))
```

```python
plot_fitting_line(linear_reg_model, X_train, Y_train, "")
```
